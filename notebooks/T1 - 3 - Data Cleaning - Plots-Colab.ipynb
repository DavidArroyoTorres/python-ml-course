```python
#k = int(np.ceil(1+np.log2(2850)))
plt.hist(data[data['Churn?']=='True.']['CustServ Calls'],bins=9,alpha=0.5, label="Churn")
plt.hist(data[data['Churn?']=='False.']['CustServ Calls'], bins = 8,alpha=0.5, label="No Churn") #bins = [0,30,60,...,200]
plt.xlabel("Nº llamadas a At. Cliente")
plt.ylabel("Frecuencia")
plt.title("Histograma de llamadas a At. Cliente")
plt.legend()

def histGraphDiscr(df,varA,varB,transp):
  for i in df[varB].unique():
     plt.hist(df[df[varB]==i][varA],bins=len(df[varA].unique()),alpha=transp, label=varB+"="+str(i))
  plt.xlabel(varA)
  plt.ylabel("Frecuencia")
  plt.title("Histograma de "+varA)
  plt.legend()



def histGraphCont(df,varA,varB,transp):
  k=int(1+np.log2(len(df[varA].unique())))
  for i in df[varB].unique():
     plt.hist(df[df[varB]==i][varA],bins=k, alpha=transp, label=varB+"="+str(i))
  plt.xlabel(varA)
  plt.ylabel("Frecuencia")
  plt.title("Histograma de "+varA)
  plt.legend()
```

```python
histGraphCont(data,"Day Charge","Churn?",0.7)
histGraphCont(data,"Eve Charge","Churn?",0.7)
histGraphCont(data,"Night Charge","Churn?",0.7)
```
